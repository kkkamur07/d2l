#### Activation Functions

```python
import torch
from d2l import torch as d2l
import matplotlib.pyplot as plt
```

```python
# Settings
x = torch.arange(-8, 8, 0.1, requires_grad=True)
y_relu = torch.relu(x)

# Calculate gradient
y_relu.backward(torch.ones_like(x), retain_graph=True) # Applying gradient to y_relu

# Plotting side-by-side
fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plot ReLU function
axes[0].plot(x.detach(), y_relu.detach())
axes[0].set_xlabel('x')
axes[0].set_ylabel('ReLU(x)')
axes[0].set_title('ReLU Activation Function')
axes[0].grid(True)

# Plot Gradient of ReLU function
axes[1].plot(x.detach(), x.grad)
axes[1].set_xlabel('x')
axes[1].set_ylabel('Gradient of ReLU(x)')
axes[1].set_title('Gradient of ReLU')
axes[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
# Reset gradient
x.grad.zero_()

# sigmoid function
y_sigmoid = torch.sigmoid(x)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

# Plotting
axes[0].plot(x.detach(), y_sigmoid.detach())
axes[0].set_xlabel('x')
axes[0].set_ylabel('Sigmoid(x)')
axes[0].set_title('Sigmoid Activation Function')
axes[0].grid(True)
# Calculate gradient
y_sigmoid.backward(torch.ones_like(x), retain_graph=True) # Applying gradient to y_sigmoid
# Plot Gradient of Sigmoid function
axes[1].plot(x.detach(), x.grad)
axes[1].set_xlabel('x')
axes[1].set_ylabel('Gradient of Sigmoid(x)')
axes[1].set_title('Gradient of Sigmoid')
axes[1].grid(True)
plt.tight_layout()
plt.show()
```

```python
# Tanh
x.grad.zero_()  # Reset gradient
y_tanh = torch.tanh(x)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
# Plotting
axes[0].plot(x.detach(), y_tanh.detach())
axes[0].set_xlabel('x')
axes[0].set_ylabel('Tanh(x)')
axes[0].set_title('Tanh Activation Function')
axes[0].grid(True)

# Calculate gradient
y_tanh.backward(torch.ones_like(x), retain_graph=True) # Applying gradient to y_tanh

axes[1].plot(x.detach(), x.grad)
axes[1].set_xlabel('x')
axes[1].set_ylabel('Gradient of Tanh(x)')
axes[1].set_title('Gradient of Tanh')
axes[1].grid(True)
plt.tight_layout()
plt.show()
```

```python
# Leaky Relu

def leaky_relu(x, alpha=0.1):
    return torch.where(x > 0, x, alpha * x)

x.grad.zero_()  # Reset gradient
y_leaky_relu = leaky_relu(x)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
# Plotting
axes[0].plot(x.detach(), y_leaky_relu.detach())
axes[0].set_xlabel('x')
axes[0].set_ylabel('Leaky ReLU(x)')
axes[0].set_title('Leaky ReLU Activation Function')
axes[0].grid(True)
# Calculate gradient
y_leaky_relu.backward(torch.ones_like(x), retain_graph=True) # Applying gradient to y_leaky_relu
# Plot Gradient of Leaky ReLU function
axes[1].plot(x.detach(), x.grad)
axes[1].set_xlabel('x')
axes[1].set_ylabel('Gradient of Leaky ReLU(x)')
axes[1].set_title('Gradient of Leaky ReLU')
axes[1].grid(True)
plt.tight_layout()
plt.show()

```

```python
# Swisch
def swish(x):
    return x * torch.sigmoid(x)
x.grad.zero_()  # Reset gradient
y_swish = swish(x)
fig, axes = plt.subplots(1, 2, figsize=(10, 4))
# Plotting
axes[0].plot(x.detach(), y_swish.detach())
axes[0].set_xlabel('x')
axes[0].set_ylabel('Swish(x)')
axes[0].set_title('Swish Activation Function')
axes[0].grid(True)
# Calculate gradient
y_swish.backward(torch.ones_like(x), retain_graph=True) # Applying gradient to y_swish
# Plot Gradient of Swish function
axes[1].plot(x.detach(), x.grad)
axes[1].set_xlabel('x')
axes[1].set_ylabel('Gradient of Swish(x)')
axes[1].set_title('Gradient of Swish')
axes[1].grid(True)
plt.tight_layout()
plt.show()
```
